# Q-Value Analysis for Averaged Waveforms

In this portion of the q-value analysis, we are utilizing the averaged neutron and photon waveforms. The data processing was done in ROOT. First, the individual events for tagged as either neutron or photon. Then each event was added as an entry in a 2-D histogram. There were two histograms, one for neutron and one for photon events. The waveforms we used are the X-profiles of these histograms. 

To calculate the q-value of the waveforms, we compare the integral over a "tail region" in the waveform to the integral over the total waveform. In this Notebook, we seek to understand the ideal range for the tail regioon. That is, the range that allows the maximum difference between the neutron and photon q-values.

```python
from matplotlib import pyplot as plt
import numpy as np
import rootpy as rt
import root_numpy as rnp

from rootpy.plotting import Hist
from rootpy.io import root_open
```

## Importing the Data from ROOT

With the code below, we import the waveform file as a .root file and then read in the two average waveforms into numpy arrays. Each waveform is split into two arrays. One contains the bin values for each entry, and the other has the actual histogram values. 

```python
data_path = "../../data/average_waveforms/"
data_tag = "take103"

avg_waveforms = root_open(data_path + data_tag + "_average_waveforms.root", 'read')

neutron_waveform, neutron_bins  = rnp.hist2array(Hist(avg_waveforms.hist_n_pfx),
                                                 return_edges = True)
photon_waveform,  photon_bins   = rnp.hist2array(Hist(avg_waveforms.hist_p_pfx), 
                                                 return_edges = True)

neutron_bins = (neutron_bins[0][:-1] + neutron_bins[0][1:]) / 2
photon_bins  = (photon_bins[0][:-1]  + photon_bins[0][1:] ) / 2
```

Below, the average neutron and average photon waveforms have been plotted together.

```python
plt.figure(figsize = (12,8))
plt.plot(neutron_bins, neutron_waveform, color = 'r')
plt.plot(photon_bins,  photon_waveform, color = 'b')

plt.legend(["Neutron Waveform", "Photon Waveform"], fontsize = 15)
plt.ylim([-0.1, 1])

plt.grid()
plt.show()
```

The following functions are used to integrate and to calculate the q-values. Additionally, the total waveform integrals are calculated so that they may be used in the subsequent analyses without wasting the time to compute the integrals repeatedly.

```python
def get_bin_index(bins, value):
    return (np.abs(bins - value)).argmin()

def integrate_waveform(start, end, bins, waveform):
    first_index = get_bin_index(bins, start)
    last_index  = get_bin_index(bins, end)
    
    # Reverse the first and the last index if the order is incorrect
    if first_index > last_index:
        first_index, last_index = last_index, first_index
        
    total = np.sum(waveform[first_index:last_index+1])
    return total

def calculate_qvalue(bins, waveform, tail_start, tail_end, total_integral):
    # Return zero if the tail region has zero extent, otherwise we'd divide by zero
    if tail_start == tail_end:
        return np.nan
    
    tail_integral = integrate_waveform(tail_start, tail_end, bins, waveform)
    qvalue = tail_integral / total_integral
    return qvalue

# The integration of the whole waveforms
INT_RANGE_START = 0
INT_RANGE_END   = 200
TOTAL_NEUTRON_INTEGRAL = integrate_waveform(INT_RANGE_START, INT_RANGE_END, neutron_bins, neutron_waveform)
TOTAL_PHOTON_INTEGRAL  = integrate_waveform(INT_RANGE_START, INT_RANGE_END, photon_bins,  photon_waveform)
```

```python
print(TOTAL_NEUTRON_INTEGRAL)
```

## Varying the Start of the Tail Region

In the first piece of the analysis, we will vary the start of the tail region while keeping the end of the tail region constant. The start of the tail ranges from the very beginning at 0 ns to 100 ns. with the end of the tail kept at 200 ns. 

```python
tail_starts = np.arange(INT_RANGE_START, 100,0.5)

neutron_qvalues = list()
photon_qvalues  = list()

for start in tail_starts:
    neutron_qvalue = calculate_qvalue(neutron_bins, neutron_waveform, 
                                      start, INT_RANGE_END, 
                                      TOTAL_NEUTRON_INTEGRAL)
    
    photon_qvalue  = calculate_qvalue(photon_bins,  photon_waveform, 
                                      start, INT_RANGE_END, 
                                      TOTAL_PHOTON_INTEGRAL)
    
    neutron_qvalues.append(neutron_qvalue)
    photon_qvalues.append(photon_qvalue)
    
neutron_qvalues = np.array(neutron_qvalues)
photon_qvalues  = np.array(photon_qvalues)
```

```python
qvalue_differences = np.abs(neutron_qvalues   - photon_qvalues)
qvalue_averages    = neutron_qvalues/2 + photon_qvalues/2

fig, axes = plt.subplots(2,1,figsize=(14,10), sharex = True)
axes[0].plot(tail_starts, neutron_qvalues, color = 'r', label = "Neutron Q-Values")
axes[0].plot(tail_starts, photon_qvalues, color = 'b', label = "Photon Q-Values")
axes[0].plot(tail_starts, qvalue_differences/qvalue_averages, color = 'g', label = "Percent Difference")
axes[0].set_title("Neutron and Photon Q-Values with Varied Start of Tail Region", fontsize = 20)
axes[0].set_xlabel("Start of the Tail Integration Region", fontsize = 15)
axes[0].set_ylabel("Q-Value", fontsize = 15)

axes[0].axvline(x = 200, color = 'k', label = "End of the Tail Region")

axes[1].plot(neutron_bins, neutron_waveform, color = 'r')
axes[1].plot(photon_bins,  photon_waveform, color = 'b')

for ax in axes:
    ax.grid()
    ax.legend(fontsize = 15)
    ax.axvline(x = 200, color = 'k', label = "End of the Tail Region")

plt.show()
```

## Varying the End of the Tail Region

Next, we will vary the ending point of the tail region while keeping the same starting position. The start of the region is kept at the 0 ns position while the end is varied from 20 ns to the end of the waveforms. 

```python
tail_ends = np.arange(20, INT_RANGE_END, 0.5)

neutron_qvalues = list()
photon_qvalues  = list()

for end in tail_ends:
    neutron_qvalue = calculate_qvalue(neutron_bins, neutron_waveform, 
                                      INT_RANGE_START, end,
                                      TOTAL_NEUTRON_INTEGRAL)
    
    photon_qvalue  = calculate_qvalue(photon_bins,  photon_waveform, 
                                      INT_RANGE_START, end,
                                      TOTAL_PHOTON_INTEGRAL)
    
    neutron_qvalues.append(neutron_qvalue)
    photon_qvalues.append(photon_qvalue)
    
neutron_qvalues = np.array(neutron_qvalues)
photon_qvalues  = np.array(photon_qvalues)
```

```python
qvalue_differences = np.abs(neutron_qvalues   - photon_qvalues)
qvalue_averages    = neutron_qvalues/2 + photon_qvalues/2

fig, axes = plt.subplots(2,1,figsize=(14,10), sharex = True)
axes[0].plot(tail_ends, neutron_qvalues, color = 'r', label = "Neutron Q-Values")
axes[0].plot(tail_ends, photon_qvalues, color = 'b', label = "Photon Q-Values")
axes[0].plot(tail_ends, qvalue_differences/qvalue_averages, color = 'g', label = "Percent Difference")
axes[0].set_title("Neutron and Photon Q-Values with Varied End of Tail Region", fontsize = 20)
axes[0].set_xlabel("End of the Tail Integration Region", fontsize = 15)
axes[0].set_ylabel("Q-Value", fontsize = 15)

axes[1].plot(neutron_bins, neutron_waveform, color = 'r', label = "Neutron Waveform")
axes[1].plot(photon_bins,  photon_waveform, color = 'b', label = "Photon Waveform")

for ax in axes:
    ax.grid()
    ax.axvline(x = 0, color = 'k', label = "Start of the Tail Region")
    ax.legend(fontsize = 15)

plt.show()
```

## Varying the Tail Region's Start *and* End Positions

```python
tail_start_steps = 100
tail_end_steps   = 100

tail_starts = np.linspace(INT_RANGE_START, INT_RANGE_END, tail_start_steps)
tail_ends   = np.linspace(INT_RANGE_START, INT_RANGE_END, tail_end_steps)

# start_end_array = np.empty((tail_start_steps, tail_end_steps), object)
neutron_qvalue_array = np.empty(shape = (tail_start_steps, tail_end_steps), dtype = float)
photon_qvalue_array  = np.empty(shape = (tail_start_steps, tail_end_steps), dtype = float)
start_ends  = np.empty(shape = (tail_start_steps, tail_end_steps), dtype = object)


for i in range(tail_start_steps):
    for j in range(tail_end_steps):
        neutron_qvalue_array[i][j] = calculate_qvalue(neutron_bins, neutron_waveform,
                                                      tail_starts[i], tail_ends[j],
                                                      TOTAL_NEUTRON_INTEGRAL)
        photon_qvalue_array[i][j]  = calculate_qvalue(photon_bins, photon_waveform,
                                                      tail_starts[i], tail_ends[j],
                                                      TOTAL_PHOTON_INTEGRAL)
        
        start_ends[i][j] = (tail_starts[i], tail_ends[j])
        
# neutron_qvalue_array = np.transpose(neutron_qvalue_array)     
# photon_qvalue_array  = np.transpose(photon_qvalue_array)
# start_ends = np.transpose(start_ends)
```

```python
epsilon = 0.5

qvalue_array_differences = np.abs(neutron_qvalue_array   - photon_qvalue_array)
qvalue_array_averages    = (neutron_qvalue_array+epsilon)/2 + (photon_qvalue_array+epsilon)/2

qvalue_array_percent_differences = qvalue_array_differences/qvalue_array_averages 
qvalue_array_percent_differences *= epsilon
```

```python
plt.figure(figsize = (15,8))
plt.imshow(qvalue_array_percent_differences, origin = 'lower',
           extent=(INT_RANGE_START, INT_RANGE_END, INT_RANGE_START, INT_RANGE_END))

plt.gray()
plt.colorbar()
plt.cm.get_cmap().set_bad(color='red')

plt.title("Absolute Percent Difference between Neutron and Photon Q-Values", fontsize = 15)
plt.xlabel("Start of the Tail Region", fontsize = 12)
plt.ylabel("End of the Tail Region",fontsize = 12)
plt.show()
```

```python
level_curve_values = np.linspace(np.nanmin(qvalue_array_percent_differences),
                                 np.nanmax(qvalue_array_percent_differences),
                                 10)

plt.figure(figsize = (15,8))

plt.imshow(qvalue_array_percent_differences, origin = 'lower',
           extent=(INT_RANGE_START, INT_RANGE_END, INT_RANGE_START, INT_RANGE_END))
plt.gray()
plt.cm.get_cmap().set_bad(color='red')

plt.title("Absolute Percent Difference between Neutron and Photon Q-Values", fontsize = 15)
plt.xlabel("Start of the Tail Region", fontsize = 12)
plt.ylabel("End of the Tail Region",fontsize = 12)
plt.colorbar()

plt.contour(qvalue_array_percent_differences, levels = level_curve_values, colors = 'g', origin = 'lower', 
           extent=(INT_RANGE_START, INT_RANGE_END, INT_RANGE_START, INT_RANGE_END))

plt.show()
```
